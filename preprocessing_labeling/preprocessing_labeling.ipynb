PREPROCESSING

```python
dataframe = pd.read_csv('data_2.csv')
```

```python
# Split the text column by newline
dataframe['comment'] = dataframe['comment'].str.split('\n')

# Join the text column back by newline
dataframe['comment'] = dataframe['comment'].str.join(' ')
```

```python
dataframe['comment'].values[30]
```

```python
dataframe['comment'] = dataframe['comment'].apply(lambda x: ' '.join(x.lower() for x in x.split()))
```

```python
dataframe["comment"] = dataframe["comment"].str.replace('\d','')
```

```python
dataframe["comment"] = dataframe["comment"].str.replace(r'[^\w\s]', ' ')
```

```python
dataframe['comment'] = dataframe['comment'].str.strip().replace('\s+', ' ', regex=True)
```
